```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.misc
import skimage.transform
plt.rcParams['figure.figsize'] = [8,8]
plt.rcParams.update({'font.size': 18})

n = 1000
q = int(n/4)
X = np.zeros((n,n))
X[(q-1):(3*q),(q-1):(3*q)] = 1

plt.imshow(X)
plt.set_cmap('gray')
plt.axis('off')
plt.title('Unrotated Matrix')
plt.show()
```

```python
X_rot = skimage.transform.rotate(X, 10)
X_rot[np.nonzero(X_rot)] = 1

plt.imshow(X_rot)
plt.set_cmap('gray')
plt.axis('off')
plt.title('Rotated Matrix')
plt.show()
```

```python
U, S, VT = np.linalg.svd(X, full_matrices=False)

plt.semilogy(S, '-o', color='k')
plt.ylim((10**(-16),10**(4)+1))
plt.yticks(np.power(10,np.arange(-16,4,4, dtype=float)))
plt.xticks(np.arange(0,1000,250))
plt.grid()
plt.title('Unrotated Matrix: Spectrum')
plt.show()
```

```python
U_rot, S_rot, VT_rot = np.linalg.svd(X_rot, full_matrices=False)

plt.semilogy(S_rot,'-o',color='k')
plt.ylim((10**(-16),10**(4)+1))
plt.yticks(np.power(10,np.arange(-16,4,4, dtype=float)))
plt.xticks(np.arange(0,1000,250))
plt.grid()
plt.title('Rotated Matrix: Spectrum')
plt.show()
```

## Result:
Here we can see that with the original image almost all of image data is captured with r=1 SVD where over 250 singular values have to be stored to store most of the information in the rotated image.

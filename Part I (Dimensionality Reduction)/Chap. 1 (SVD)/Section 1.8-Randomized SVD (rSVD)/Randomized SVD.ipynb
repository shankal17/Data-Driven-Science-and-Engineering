```python
import matplotlib.pyplot as plt
from matplotlib.image import imread
import numpy as np
import os

plt.rcParams['figure.figsize'] = [32, 12]
plt.rcParams.update({'font.size':18})

# Define randomized SVD function
def rSVD(X, r, q, p):
    # Step 1: Sample column space of X with P matrix
    ny = X.shape[1]
    P = np.random.randn(ny, r+p)
    Z = X @ P
    for k in range(q):
        Z = X @ (X.T @ Z)
        
    Q, R = np.linalg.qr(Z, mode='reduced')
    
    # Step 2: Compute SVD on projected Y = Q.T @ X
    Y = Q.T @ X
    Uy, S, VT = np.linalg.svd(Y, full_matrices=False)
    U = Q @ Uy
    return U, S, VT
```

```python
A = imread('jupiter.jpg')
X = np.mean(A,axis=2) # Convert RGB -> grayscale

U, S, VT = np.linalg.svd(X,full_matrices=0) # Deterministic SVD

r = 400 # Target rank
q = 1   # Power iterations
p = 5   # Oversampling parameter

rU, rS, rVT = rSVD(X,r,q,p)
```

```python
## Reconstruction
XSVD = U[:,:(r+1)] @ np.diag(S[:(r+1)]) @ VT[:(r+1),:] # SVD approximation
errSVD = np.linalg.norm(X-XSVD,ord=2) / np.linalg.norm(X,ord=2)

XrSVD = rU[:,:(r+1)] @ np.diag(rS[:(r+1)]) @ rVT[:(r+1),:] # SVD approximation
errSVD = np.linalg.norm(X-XrSVD,ord=2) / np.linalg.norm(X,ord=2)
```

```python
## Plot
fig, axs = plt.subplots(1,3)

title_size = 40
plt.set_cmap('gray')
axs[0].imshow(X)
axs[0].axis('off')
axs[0].set_title('Original Image', fontsize=title_size)
axs[1].imshow(XSVD)
axs[1].axis('off')
axs[1].set_title('Deterministic SVD Image', fontsize=title_size)
axs[2].imshow(XrSVD)
axs[2].axis('off')
axs[2].set_title('Randomized SVD Image', fontsize=title_size)

plt.show()
```

## Results:
Comparing the images above, we can see that both the deterministic SVD and the randomized SVD accurately capture the original image. The randomized SVD however is much faster and less computationally demanding.

```python
## Illustrate power iterations
X = np.random.randn(1000,100)
U, S, VT = np.linalg.svd(X,full_matrices=0)
S = np.arange(1,0,-0.01)
X = U @ np.diag(S) @ VT

color_list = np.array([[0,0,2/3],  # Define color map
                 [0,0,1],
                 [0,1/3,1],
                 [0,2/3,1],
                 [0,1,1],
                 [1/3,1,2/3],
                 [2/3,1,1/3],
                 [1,1,0],
                 [1,2/3,0],
                 [1,1/3,0],
                 [1,0,0],
                 [2/3,0,0]])

plt.plot(S,'o-',color='k',LineWidth=2,label='SVD')

Y = X
for q in range(1,6):
    Y = X.T @ Y
    Y = X @ Y
    Uq, Sq, VTq = np.linalg.svd(Y,full_matrices=0)
    plt.plot(Sq,'-o',color=tuple(color_list[2*q+1]),LineWidth=2,label='rSVD, q = '+str(q))

plt.legend()
plt.show()
```

## Result:
Here we can see that the singular values of the matrix drop off more and more quickly with increasing power iterations.
